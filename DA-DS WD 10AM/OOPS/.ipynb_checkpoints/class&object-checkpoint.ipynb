```python
def stripped(string):
    start = 0
    end = len(string) - 1

    for i in range(len(string)):
        if string[i] != ' ':
            start = i
            break
    for j in range(end, -1, -1):
        if string[j] != ' ':
            end = j
            break
    return string[start : end + 1]
```

```python
stripped('    Tushar   khitoliya    ')
```

```python
for j in range(10, -1, -1):
    print(j)
```

```python
def splitting(string):
    string = stripped(string)
    string = string + ' '
    a = []
    idx = 0
    for i in range(len(string)):
        if string[i] == " ":
            a.append(string[idx:i])
            idx = i + 1
    b = []
    for j in a:
        if j != '':
            b.append(j)
    return b
```

```python
splitting('   training       basket  ')
```

```python
class Atm:
    def __init__(self): # constructor
        self.pin = None
        self.balance = None
        print(f"Welcome to ICICI Bank, How may i assist you : ")
        self.menu()
    def menu(self):
        print('''PRESS 1. TO SET PIN
                \n PRESS 2. TO CHANGE PIN
                \n PRESS 3. TO CHECK BALANCE
                \n PRESS 4. TO WITHDRAW MONEY
                \n PRESS ANYTHING.''')
        user_input = input("enter the service you want to use : ")
        if user_input == '1':
            self.set_pin()            
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw_money()
        else:
            print("thank you")
    
    def set_pin(self):
        user_pin = input("enter your new pin : ")
        user_balance = float(input("enter your balance : "))
        self.pin = user_pin
        self.balance = user_balance
        print(f"Your Pin has been generated successfully")
        self.menu()

    def change_pin(self):
        user_pin = input("enter your previous pin : ")
        if user_pin == self.pin:
            new_pin = input("enter the new pin : ")
            self.pin = new_pin
            print(f"Your Pin has been changed successfully")
        else:
            print(f"yaha se bhag jao ")
        self.menu()

    def check_balance(self):
        user_pin = input("enter your current pin : ")
        if user_pin == self.pin:
            print(f"Your balance is ₹{self.balance}")
        else:
            print(f"chor sala")
        self.menu()

    def withdraw_money(self):
        user_pin = input("enter your current pin : ")
        if user_pin == self.pin:
            amount = int(input("enter the amount you want to withdraw : "))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print(f"Your balance is ₹{self.balance}")
            else:
                print(f"Insufficient Balance")
        else:
            print(f"chor sala")
        self.menu()
```

```python
class Fraction:
    def __init__(self, x, y): # parameterized constructor
        self.num = x
        self.den = y

    def __str__(self):
        return f"Fraction of {self.num, self.den} = {self.num}/{self.den}"

    def addition(self, x1, y1):
        self.num1 = x1
        self.den1 = y1
        new_num = self.num1*self.den + self.num*self.den1
        new_den = self.den*self.den1
        return f"Addition = {new_num}/{new_den}"
```
