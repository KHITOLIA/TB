```python
class Carr:
    __no_instance = 0
    def __init__(self, name, year):
        self.name = name
        self.year = year
        Carr.__no_instance += 1

    def get_instance():
        return Carr.__no_instance
```

```python
bmw = Carr('M5', '2024')
```

```python
bmw._Carr__no_instance
```

```python
Carr.get_instance()
```

```python
maruti = Carr('victoris', 2025)
```

```python
Carr.get_instance()
```

# Inheritence

```python
# Parent
class User:
    def __init__(self, age):
        self.name = 'Tushar'
        self.age = age
    def login(self):
        print("login")
        
# Child
class Student(User):
    def __init__(self, age_s):
        self.age_s = age_s
    def login(self):
        print("hello")
```

```python
u = User(23)
```

```python
s = Student(29)
```

```python
s.login()
```

```python
class Device:
    def __init__(self, name, price):
        self.__name = name
        self.__price = price
        print("Device created")

    def display(self):
        print(self.name, self.price)
```

```python
class Phone(Device):
    def __init__(self, name,price):
        super().__init__(name, price)
        self.price = price
        self.name = name
    def display(self):
        super().display()
        print(f"phone specs {self.name}")
```

```python
p = Phone('samsung', 30000)
```

```python
p.display()
```
