##Q-1: Rectangle Class

1. Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.

 

2. Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of the rectangle.

 

3. Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.

```python
class Rectangle:
    def __init__(self, l, w): # constructor
        self.l = l
        self.w = w

    def perimeter(self):
        print(f"Perimeter of this rectangle {2*(self.l + self.w)}cm")

    def area(self):
        print(f"Area of this rectangle {self.l * self.w}cm^2")

    def display(self):
        print(f"Length of Rectangle {self.l}")
        print(f"Width of Rectangle {self.w}")
        print(" ")
        self.perimeter()
        print(" ")
        self.area()
        
```

```python
a = Rectangle(3, 2)
```

```python
a.perimeter()
```

```python
a.area()
```

```python
a.display()
```

```python
a.w
```
