# Write OOP classes to handle the following scenarios:
1. - A user can create and view 2D coordinates
2. - A user can find out the distance between 2 coordinates
3. - A user can find find the distance of a coordinate from origin
4. - A user can check if a point lies on a given line
5. - A user can find the distance between a given 2D point and a given line



```python
class Coordinate:
    def __init__(self, x1, y1):
        self.x1 = x1
        self.y1 = y1
    def __str__(self):
        return f"Coordinates are {self.x1, self.y1}"

    def distance_between_point(self, x2, y2):
        self.x2 = x2
        self.y2 = y2
        dist = ((self.x2 - self.x1)**2 + (self.y2 - self.y1)**2)**0.5
        return f"Distance between {(self.x1, self.y1) and (self.x2, self.y2)} = {dist}", (self.x2, self.y2)

    def equation_of_line(self, x2, y2):
        self.x2 = x2
        self.y2 = y2

        m = (self.y2 - self.y1)/(self.x2 - self.x1)
        c = self.y2 - m*self.x2
        x3 = int(input("enter the x coordinate : "))
        y3 = int(input("enter the y coordinate : "))
        print(f"y = {m}*x + {c}")

        if y3-m*x3-c == 0:
            print(f"This point {x3, y3} lies on the line")
        else:
            dist = y3-m*x3-c
            print(f"Distance between line and point is {dist}")
```

```python
class Atm:

    __counter = 0
    
    def __init__(self): # constructor
        self.__pin = None
        self.__balance = None
        print(f"Welcome to ICICI Bank, How may i assist you : ")
        self.cid = Atm.__counter + 1
        Atm.__counter = Atm.__counter + 1
        print(f"Customer ID : {self.cid}")
        self.menu()

    @staticmethod
    def get_counter():
        return  Atm.__counter
        
    def menu(self):
        print('''PRESS 1. TO SET PIN
                \n PRESS 2. TO CHANGE PIN
                \n PRESS 3. TO CHECK BALANCE
                \n PRESS 4. TO WITHDRAW MONEY
                \n PRESS ANYTHING.''')
        user_input = input("enter the service you want to use : ")
        if user_input == '1':
            self.set_pin()            
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw_money()
        else:
            print("thank you")
    
    def set_pin(self):
        user_pin = input("enter your new pin : ")
        user_balance = float(input("enter your balance : "))
        self.__pin = user_pin
        self.__balance = user_balance
        print(f"Your Pin has been generated successfully")
        self.menu()

    def change_pin(self):
        user_pin = input("enter your previous pin : ")
        if user_pin == self.__pin:
            new_pin = input("enter the new pin : ")
            self.__pin = new_pin
            print(f"Your Pin has been changed successfully")
        else:
            print(f"yaha se bhag jao ")
        self.menu()

    def check_balance(self):
        user_pin = input("enter your current pin : ")
        if user_pin == self.__pin:
            print(f"Your balance is ₹{self.__balance}")
        else:
            print(f"chor sala")
        self.menu()

    def withdraw_money(self):
        user_pin = input("enter your current pin : ")
        if user_pin == self.__pin:
            amount = int(input("enter the amount you want to withdraw : "))
            if amount <= self.__balance:
                self.__balance = self.__balance - amount
                print(f"Your balance is ₹{self.__balance}")
            else:
                print(f"Insufficient Balance")
        else:
            print(f"chor sala")
        self.menu()
```
