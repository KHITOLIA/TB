```python
def stripped(string):
    start = 0
    end = len(string) - 1

    for i in range(len(string)):
        if string[i] != ' ':
            start = i
            break
    for j in range(end, -1, -1):
        if string[j] != ' ':
            end = j
            break
    return string[start : end + 1]
```

```python
stripped('    Tushar   khitoliya    ')
```

```python
for j in range(10, -1, -1):
    print(j)
```

```python
def splitting(string):
    string = stripped(string)
    string = string + ' '
    a = []
    idx = 0
    for i in range(len(string)):
        if string[i] == " ":
            a.append(string[idx:i])
            idx = i + 1
    b = []
    for j in a:
        if j != '':
            b.append(j)
    return b
```

```python
splitting('   training       basket  ')
```

```python
class Atm:

    def __init__(self):
        self.pin = None
        self.balance = None
        print(f"Welcome to ICICI Bank, How may i assist you : ")
        
    def menu(self):
        print('''PRESS 1. TO SET PIN
                \n PRESS 2. TO CHANGE PIN
                \n PRESS 3. TO CHECK BALANCE
                \n PRESS 4. TO WITHDRAW MONEY
                \n PRESS ANYTHING.''')
        user_input = input("enter the service you want to use : ")
        if user_input == '1':
            self.set_pin()            
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            print("check balance")
        elif user_input == '4':
            print("withdraw money")
        else:
            print("thank you")
    
    def set_pin(self):
        user_pin = input("enter your new pin : ")
        user_balance = float(input("enter your balance : "))
        self.pin = user_pin
        self.balance = user_balance
        print(f"Your Pin has been generated successfully")

    def change_pin(self):
        user_pin = input("enter your previous pin : ")
        if user_pin == self.pin:
            new_pin = input("enter the new pin : ")
            self.pin = new_pin
            print(f"Your Pin has been changed successfully")
        else:
            print(f"yaha se bhag jao ")
```
