```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Relational plot
1. scatterplot
2. lineplot

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level f(x)
import warnings
warnings.filterwarnings('ignore')

# sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
sns.lineplot(data = tips, x = 'time', y = 'tip', hue = 'sex', style = 'smoker', ci = None)
plt.show()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size', kind = 'scatter')
```

```python
ins = pd.read_csv('insurance.csv')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line', col = 'sex', row = 'day')
```

# Distribution plot
1. histplot
2. kdeplot
3. distplot
4. displot

```python
sns.histplot(data  = tips, x = 'total_bill', bins = 5)
```

```python
sns.kdeplot(data  = tips, x = 'total_bill')
```

```python
sns.distplot(x = tips['total_bill'])
```

```python
sns.distplot(x = tips['tip'], hist = True, kde = False)
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'kde', hue = 'sex' , fill = True)
```

```python
num = [col for col in tips.columns if tips[col].dtypes != 'category']
```

```python
tips.info()
```
