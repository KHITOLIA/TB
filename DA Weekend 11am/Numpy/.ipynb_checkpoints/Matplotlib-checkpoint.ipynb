```python
# univariate analysis : Numerical and categorical

numerical = []
categorical = []

for col in ins.columns:
    if ins[col].dtypes == object:
        categorical.append(col)
    else:
        numerical.append(col)
```

```python
def univariate_analysis_numerical(data):
    for col in numerical:
        plt.hist(data[col], bins=20)
        plt.title(f'Distribution of {col}')
        plt.grid()
        plt.show()

univariate_analysis_numerical(ins)
```

```python
ins.charges.mean() - ins.charges.median()
```

```python
ins.charges.skew()
```

```python
def univariate_analysis_categorical(data):
    for col in categorical:
        plt.bar(ins[col].value_counts().reset_index()[col], ins[col].value_counts().reset_index()['count'])
        plt.title(f'Count per {col}')
        plt.show()

univariate_analysis_categorical(ins)
```
