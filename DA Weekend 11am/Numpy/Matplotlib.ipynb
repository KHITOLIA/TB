```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Line plot
```

```python
x = np.random.randint(0, 100, 100) 
```

```python
x
```

```python
y = 2*x + 1
```

```python
len(x) == len(y)
```

```python
plt.plot(x)
```

```python
plt.plot(y)
```

```python
plt.plot(y, x, marker = '*', linewidth = 5, label = 'Y')
plt.legend()
plt.title('Y vs X graph')
plt.show()
```

```python
insurance = pd.read_csv("insurance.csv")
```

```python
age_female = insurance.groupby('sex').get_group('female')['age']
```

```python
age_female
```

```python
age_male = insurance.groupby('sex').get_group('male')['age']
```
