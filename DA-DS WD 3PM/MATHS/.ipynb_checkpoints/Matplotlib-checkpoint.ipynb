```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
# line plot : used to check the relationship between two features or trend analysis
```

```python
plt.figure(figsize=(40, 5))
plt.plot(df['Age'], df['Pregnancies'])
```

```python
plt.figure(figsize=(40, 5))
plt.scatter(df['Age'], df['Pregnancies'], label = 'Age vs Pregnancies')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Pregnancies')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.scatter(df.groupby('Outcome').get_group(1)['Age'], df.groupby('Outcome').get_group(1)['Pregnancies'], label = 'Diabetic Patients')
plt.legend()

plt.subplot(2, 1, 2)
plt.scatter(df.groupby('Outcome').get_group(0)['Age'], df.groupby('Outcome').get_group(0)['Pregnancies'], label = 'Non - Diabetic Patients')
plt.legend()

plt.show()
```
