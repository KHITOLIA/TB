```python
# Univariate Analysis
def univariate_analysis_numerical(insurance):
    i = 1
    for col in numerical_feat:    
        plt.figure(figsize=(10, 8))
        plt.subplot(4, 1, i)
        i = i + 1
        sns.distplot(insurance[col])
        plt.title(f"{col} Distribution")
        if insurance[col].mean() - insurance[col].median() > 0:
            print(f"the {col} distribution is right skewed")
        else:
            print(f"the {col} distribution is left skewed")
        plt.show()
        
```

```python
univariate_analysis_numerical(insurance)
```

```python
categorical_feat
```

```python
ax = sns.countplot(x = insurance['region'])
for cont in ax.containers:
    ax.bar_label(cont)
```

```python
def univariate_analysis_categorical(insurance):
    i = 1
    for col in categorical_feat:
        plt.figure(figsize=(5, 10))
        plt.subplot(len(categorical_feat), 1, i)
        i = i + 1
        ax = sns.countplot(x = insurance[col], palette='magma')
        for cont in ax.containers:
            ax.bar_label(cont)
        plt.title(f"{col} Value count per Category")
        # plt.tight_layout()
        plt.xticks(rotation = 90)
```

```python
univariate_analysis_categorical(insurance)
```

```python
def bivariate_analysis_numerical(insurance):
    l = numerical_feat
    # l.remove('charges')
    i = 1
    for col in l:
        plt.figure(figsize=(20, 10))
        plt.subplot(len(l), 1, i)
        i = i + 1
        sns.scatterplot(x = col, y = 'charges', data = insurance)
        plt.title(f"{col} V/s Insurance Charge")
```

```python
bivariate_analysis_numerical(insurance)
```

```python
def bivariate_analysis_categorical(insurance):
    i = 1
    for col in categorical_feat:
        plt.figure(figsize=(10, 8))
        plt.subplot(len(categorical_feat), 1, i)
        i = i + 1
        sns.lineplot(x = col, y = 'charges', data = insurance, ci = None)
        plt.title(f"{col} V/s Insurance Charge")
```
