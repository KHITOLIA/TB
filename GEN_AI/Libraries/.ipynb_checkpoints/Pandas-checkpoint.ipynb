```python
ipl['Team1Players'].iloc[0].replace("[", "").split(",")[0]
```

```python
team1openers = []
team2openers = []
for i in range(ipl.shape[0]):
    team1openers.append(ipl['Team1Players'].iloc[i].replace("[", "").split(",")[0] + '/' + ipl['Team1Players'].iloc[i].replace("[", "").split(",")[1])
    team2openers.append(ipl['Team2Players'].iloc[i].replace("[", "").split(",")[0] + '/' + ipl['Team2Players'].iloc[i].replace("[", "").split(",")[1])
```

```python
ipl['team1openers'] = team1openers
ipl['team2openers'] = team2openers
```

```python
def missing_columns(data):
    mis_col = []
    for col in data.columns:
        if data[col].isnull().sum() != 0:
            mis_col.append(col)
    return mis_col
```

```python
mis_col = missing_columns(ipl)
mis_col
```

```python
def unique_values(data, mis_col):
    for col in mis_col:
        print(data[col].unique)
```

```python
ipl['SuperOver'] = ipl['SuperOver'].fillna(ipl['SuperOver'].value_counts().reset_index()['SuperOver'][0])
```

```python
ipl['WinningTeam'] = ipl['WinningTeam'].fillna(ipl['WinningTeam'].value_counts().reset_index()['WinningTeam'][0])
```

```python
ipl['Player_of_Match'] = ipl['Player_of_Match'].fillna(ipl['Player_of_Match'].value_counts().reset_index()['Player_of_Match'][0])
```

```python
insurance.groupby(['smoker', 'region', 'sex']).agg({
    'age' : 'median',
    'region': 'count',
    'bmi' : 'median',
    'charges' : 'mean',
    'smoker' : 'count'
})
```
