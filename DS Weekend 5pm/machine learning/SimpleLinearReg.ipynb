```python
m = 0
b = 0
def personal_linear_regression(x_train, y_train):
    print(x_train.shape, y_train.shape)
    num = 0
    den = 0
    for i in range(x_train.shape[0]):
        num +=  (x_train[i] - np.mean(x_train)) * (y_train[i] - np.mean(y_train))
        den += (x_train[i] - np.mean(x_train))**2
    m = num/den
    print(m)
    b = np.mean(y_train) - m*np.mean(x_train)
    print(b)
    return m, b

```

```python
m, b = personal_linear_regression(x_train, y_train)
```
