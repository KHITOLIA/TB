```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=60, random_state=21)
```

```python
plt.scatter(x, y)
```

```python
# lets apply ols and find m ,b
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
m = model.coef_
b = model.intercept_
b, m
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color = 'black')
```

```python
# lets apply gradient descent for m = 32.65469443 b = -100
m = 32.65469443
b = -100
y_pred = m*x + b
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color = 'black')
plt.plot(x, y_pred, color = 'red', label = 'b = -100')
```

```python
# lets take learning rate = 0.1
lr = 0.01
# slope at b = 0
slope = -2*(np.sum(y - m*x.ravel() - b))

step_size = lr*slope
print(step_size, slope) # at b = 0

# Updating the new intercept
b = b -step_size
b
```

```python
y_pred_1 = m*x + b
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color = 'orange')
plt.plot(x, y_pred, color = 'red', label = 'b = -100')
plt.plot(x, y_pred_1, color = 'green', label = 'new b')
```
