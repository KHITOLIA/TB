```python
x = np.arange(0, 10, 1)
```

```python
x
```

```python
y = 2*x**2 + 3
```

```python
y
```

```python
# plt.figure(figsize=(20, 6))
plt.subplot(3, 1, 1)
plt.plot(x, y, c = 'red', label = 'Y vs X', linewidth = 4, linestyle = '--', marker = 'o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()

plt.subplot(3, 1, 2)
plt.plot(x, c= 'green', label = 'x')
plt.subplot(3, 1, 3)
plt.plot(y, c = 'black', label = 'y')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

# plt.xlabel('X-Axis')
# plt.ylabel('Y-Axis')
# plt.legend()
# plt.show()
```

```python
singers = ['arijit', 'sidho', 'bpraak', 'karan', 'masoom']
pop = [100, 90, 70, 110, 90]
```

```python
plt.bar(singers, pop, color = 'red')
plt.xlabel('Singers')
plt.ylabel('Popularity')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.pie(x = pop, labels = singers, explode = [0.1, 0.2, 0.5, 0.4, 0.01], shadow = True, autopct = "%.4f")
plt.show()
```

```python
insurance = pd.read_csv("insurance.csv")
```

```python
insurance.groupby('sex').get_group('female')['bmi']
```

```python
plt.figure(figsize=(30, 4))
plt.plot(insurance.groupby('sex').get_group('female')['bmi'], label = 'female')
plt.plot(insurance.groupby('sex').get_group('male')['bmi'], label = 'male')
plt.legend()
plt.show()
```
